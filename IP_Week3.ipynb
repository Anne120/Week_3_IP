```python
#Importing the numpy library 
import numpy as np

#Importing the pandas library 
import pandas as pd

#importing matplot library
import matplotlib as plt

#import seaborn library
import seaborn as sns
```

```python
#creating a dataframe from the description excel file
df = pd.read_excel('/content/CDR_description.xlsx')
df
```

```python

#creating a dataframe from the cells geo csv file
df1 = pd.read_csv('/content/cells_geo.csv', delimiter=';')
df1
```

```python
#creating a data frame from CDR 20120507 url
df_url = "http://bit.ly/Telcom_dataset1"
df3 = pd.read_csv(df_url)
df3

#changing the column names 

df3.columns = ['PRODUCT',	'VALUE',	'DATETIME',	'CELL_ON_SITE',	'DW_A_NUMBER_INT',	'DW_B_NUMBER_INT',	'COUNTRY_A',	'COUNTRY_B',	'CELL_ID',	'SITE_ID']
df3
```

```python
#creating a data frame from CDR 20120508 url
df_url = "http://bit.ly/Telcom_dataset2"
df4 = pd.read_csv(df_url)
df4.head()

#changing the column names
df4.columns = ['PRODUCT',	'VALUE',	'DATETIME',	'CELL_ON_SITE',	'DW_A_NUMBER_INT',	'DW_B_NUMBER_INT',	'COUNTRY_A',	'COUNTRY_B',	'CELL_ID',	'SITE_ID']
df4.head()
```

```python
#creating a data frame from CDR 20120509 url
df_url = "http://bit.ly/Telcom_dataset3"
df5 = pd.read_csv(df_url)
df5.head()

#changing the column names
df5.columns = ['PRODUCT',	'VALUE',	'DATETIME',	'CELL_ON_SITE',	'DW_A_NUMBER_INT',	'DW_B_NUMBER_INT',	'COUNTRY_A',	'COUNTRY_B',	'CELL_ID',	'SITE_ID']
df5.head()
```

```python
#Joining the three CDR datasets and calling it merge_1
df_merge1 = pd.merge(left= df4, right=df5, on=['PRODUCT',	'VALUE',	'DATETIME',	'CELL_ON_SITE',	'DW_A_NUMBER_INT',	'DW_B_NUMBER_INT',	'COUNTRY_A',	'COUNTRY_B',	'CELL_ID',	'SITE_ID'], how='left')

#dropping the duplicate values
df4.drop_duplicates(subset=['PRODUCT',	'VALUE',	'DATETIME',	'CELL_ON_SITE',	'DW_A_NUMBER_INT',	'DW_B_NUMBER_INT',	'COUNTRY_A',	'COUNTRY_B',	'CELL_ID',	'SITE_ID'], inplace=True)

#running the newly merged dataset
df_merge1
```

```python
#merging df1, and merge1 and calling it merge2
df_merge2 = pd.concat([df1, df_merge1], axis=1)
df_merge2

```

```python
#dropping duplicates in the merged table
df_merge2.drop_duplicates(inplace = True)
df_merge2
```

```python
#Checking for null values in the merged table
df_merge1.isnull()

#drop rows where data is null
df_merge1.dropna(how = 'all')
df_merge1

#dropping columns where data is null
df_merge1.dropna(axis=1, how ='all')
df_merge1
```
